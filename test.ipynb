```python
import lstm
import time
import poloniex
import pandas as pd
```

```python
filename = "USDT_BTC.csv"
polo = poloniex.Poloniex()
chart_data = polo.returnChartData('USDT_BTC',period=7200,start=time.time() - polo.DAY * 100, end=time.time())
```

```python
df = pd.DataFrame(chart_data)
df.to_csv(path_or_buf=filename, header = False, index= False)
```
